# One Hot Encodings

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample categorical data
data = np.array([['yellow'], ['Black'], ['Green'], ['Black'], ['yellow']])

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # sparse_output=False returns an array

# Fit and transform the data
encoded_data = encoder.fit_transform(data)
```

# Bag of Words (BOW)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["I love Bangladesh",
                         "Bangladesh love Bangladesh",
                         "I belong here",
                          "Bangladesh beautyful country"],"output":[1,0,0,1]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
bow
```

```python
bow.toarray()
```

```python
#vocabulary
print(cv.vocabulary_)
```

# N-Grams

```python
df = pd.DataFrame({"text":["I love Bangladesh",
                         "Bangladesh love Bangladesh",
                         "I belong here",
                          "Bangladesh beautyful country"],"output":[1,0,0,1]})

df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
bow.toarray()
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
#Ti gram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
bow.toarray()
```

```python
print(cv.vocabulary_)
```

```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
story = []
for filename in os.listdir('data'):
    if filename == '.ipynb_checkpoints':
      pass
    f = open(os.path.join('data',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vec = model.wv.get_normed_vectors()
```

```python
vec
```

```python
len(vec[0])
```
